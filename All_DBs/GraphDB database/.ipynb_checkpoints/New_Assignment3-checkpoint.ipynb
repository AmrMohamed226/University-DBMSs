```python
import csv
from rdflib import Graph, Literal, Namespace, URIRef
from rdflib.namespace import DCTERMS, RDF, RDFS, SKOS, XSD
import pandas as pd
# import rdfpandas
```

```python
Assigned_Hours=pd.read_csv('Assigned_Hours.csv')
Course_Instances=pd.read_csv('Course_Instances.csv')
Course_plannings=pd.read_csv('Course_plannings.csv')
Courses=pd.read_csv('Courses.csv')
Programme_Courses=pd.read_csv('Programme_Courses.csv')
Programmes=pd.read_csv('Programmes.csv')
Registrations=pd.read_csv('Registrations.csv')
Reported_Hours=pd.read_csv('Reported_Hours.csv')
Senior_Teachers=pd.read_csv('Senior_Teachers.csv')
Students=pd.read_csv('Students.csv')
Teaching_Assistants=pd.read_csv('Teaching_Assistants.csv')
```

```python
# Teaching Assistants


Teaching_Assistants.head()

Teaching_Assistants.rename(columns={"Teacher name": "personName", "Teacher id": "personID", "Department name": "departmentName",'Division name': "divisionName"}, inplace=True)

Teaching_Assistants['personName'] = Teaching_Assistants['personName'].str.replace(' ','-')

# make a graph
output_graph = Graph()
uri=Namespace("http://www.semanticweb.org/keyha/ontologies/2022/3/untitled-ontology-11#")
output_graph.bind("uri", uri)

# Teaching_Assistants
for index, row in Teaching_Assistants.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('TeachingAssistant') )))        
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"personName"),
                        URIRef(uri+Literal(row['personName']) )))
    
    #output_graph.add(  (URIRef(uri+row.personID),
    #                    URIRef(RDF.type),
    #                    URIRef(Literal(row['personID'], datatype=XSD.string)) ))
            
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+'departmentName'),
                        URIRef(uri+Literal(row['departmentName']) )))
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"divisionName"),
                        URIRef(uri+Literal(row['divisionName']) )))
    
```

```python
# Senior Teachers

Senior_Teachers.rename(columns={"Teacher name": "personName", "Teacher id": "personID", "Department name": "departmentName",'Division name': "divisionName"}, inplace=True)

Senior_Teachers['personName'] = Teaching_Assistants['personName'].str.replace(' ','-')

# Senior_Teachers.columns

# Senior_Teachers.head()

# uri="http://www.semanticweb.org/keyha/ontologies/2022/3/untitled-ontology-11#"

for index, row in Senior_Teachers.iterrows():
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('SeniorTeacher') ))) 

    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"personName"),
                       URIRef(uri+Literal(row['personName']) )))
    
 
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"departmentName"),
                        URIRef(uri+Literal(row['departmentName']) )))
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"divisionName"),
                        URIRef(uri+Literal(row['divisionName']) )))
    
    
    #output_graph.add(  (URIRef(uri+row.personID),
                        #URIRef(uri+"TeachBelongsDiv"),
                        #URIRef(uri+row.divisionName) ))
```

```python
# Course
Courses=pd.read_csv('Courses.csv')
Courses.head()
Courses['level'] = Courses['level'].str.replace(' ','-')
Courses.rename(columns={"Course name": "courseName", "CourseCode": "courseCode", "Credits": "credits",'level': "level", "Department": "departmentName",'Division': "divisionName",'Owned By': "programCode"}, inplace=True)


for index, row in Courses.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('Course') ))) 
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"courseName"),
                        URIRef(uri+Literal(row['courseName']) )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"credits"),
                        URIRef(uri+Literal(row['credits']) )))
       
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"level"),
                        URIRef(uri+Literal(row['level']) )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"departmentName"),
                        URIRef(uri+Literal(row['departmentName']) )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"divisionName"),
                        URIRef(uri+Literal(row['divisionName']) )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"programCode"),
                        URIRef(uri+Literal(row['programCode']) )))
```

```python
# ProgramCourses

Programme_Courses.rename(columns={"Programme code": "programCode", "Study Year": "studyYear", "Academic Year": "academicYearProgIn",'Course': "courseCode", "Course Type": "courseType"}, inplace=True)

Programme_Courses.head()

#Programme_Courses.info()

for index, row in Programme_Courses.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('ProgramCourse') )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"studyYear"),
                        URIRef(uri+Literal(row['studyYear']) )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"academicYearProgIn"),
                        URIRef(uri+Literal(row['academicYearProgIn']) )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"courseCode"),
                        URIRef(uri+Literal(row['courseCode']) )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"courseType"),
                        URIRef(uri+Literal(row['courseType']) )))
    
```

```python
# Students

Students.head()

Students.rename(columns={"Student name": "personName", "Student id": "personID", "Programme": "programCode","Year": "RegistrationYear",'Graduated':'graduated'}, inplace=True)

Students['personName'] = Students['personName'].str.replace(' ','-')

Students.head()

for index, row in Students.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('Student') )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"personName"),
                        URIRef(uri+Literal(row['personName']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"programCode"),
                        URIRef(uri+Literal(row['programCode']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"RegistrationYear"),
                        URIRef(uri+Literal(row['RegistrationYear']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"graduated"),
                        URIRef(uri+Literal(row['graduated']) )))
    
```

```python
# Reported_Hours

Reported_Hours.rename(columns={"Hours": "reportedHours", "Course code": "courseCode","Teacher Id":'personID'}, inplace=True)

Reported_Hours.head()

Reported_Hours.columns

for index, row in Reported_Hours.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('Assignment') )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"reportedHours"),
                        URIRef(uri+Literal(row['reportedHours']) )))
    
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"courseCode"),
                        URIRef(uri+Literal(row['courseCode']) )))
```

```python
# Registrations

Registrations.rename(columns={"Course Instance": "CourseInstID", "Student id": "personID", "Status": "status",'Grade': "grade"}, inplace=True)

Registrations.head()

Registrations['status'] = Registrations['status'].str.replace(' ','-')

Registrations.status.unique()

for index, row in Registrations.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('Registration') )))
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"CourseInstID"),
                        URIRef(uri+Literal(row['CourseInstID']) )))   
   
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"status"),
                        URIRef(uri+Literal(row['status']) )))
    
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(uri+"grade"),
                        URIRef(uri+Literal(row['grade']) )))
```

```python
# Programmes

Programmes.rename(columns={"Programme name": "programName", "Programme code": "programCode", "Department name": "departmentName",'Director': "personID"}, inplace=True)

Programmes.head()

for index, row in Programmes.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.programCode),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('program') )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"programName"),
                        URIRef(uri+Literal(row['programName']) )))
    
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"departmentName"),
                        URIRef(uri+Literal(row['departmentName']) )))
                            
    output_graph.add(  (URIRef(uri+str(row.programCode)),
                        URIRef(uri+"personID"),
                        URIRef(uri+Literal(row['personID']) )))
                        
```

```python
# Course_Planning

Course_plannings.rename(columns={"Course": "courseCode", "Planned number of Students": "planningNumStd", "Senior Hours": "seniorHours",'Assistant Hours': "assistantHours"}, inplace=True)

Course_plannings.head()

for index, row in Course_plannings.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.courseCode),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('CourseInstanse') )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"planningNumStd"),
                        URIRef(uri+Literal(row['planningNumStd']) )))
    
    output_graph.add(  (URIRef(uri+row.courseCode),
                        URIRef(uri+"seniorHours"),
                        URIRef(uri+Literal(row['seniorHours']) )))
    
    output_graph.add(  (URIRef(uri+str(row.courseCode)),
                        URIRef(uri+"assistantHours"),
                        URIRef(uri+Literal(row['assistantHours']) )))
```

```python
# Course_Instances

Course_Instances.rename(columns={"Course code": "courseCode", "Study period": "studyPeriod", "Academic year": "academicYearCourIn",'Instance_id': "CourseInstID",'Examiner':"personID"}, inplace=True)

Course_Instances.head()

for index, row in Course_Instances.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.courseCode),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('CourseInstanse') )))
    
    output_graph.add(  (URIRef(uri+str(row.CourseInstID)),
                        URIRef(uri+"courseCode"),
                        URIRef(uri+Literal(row['courseCode']) )))
    
    output_graph.add(  (URIRef(uri+str(row.CourseInstID)),
                        URIRef(uri+"studyPeriod"),
                        URIRef(uri+Literal(row['studyPeriod']) )))
    
    output_graph.add(  (URIRef(uri+str(row.CourseInstID)),
                        URIRef(uri+"academicYearCourIn"),
                        URIRef(uri+Literal(row['academicYearCourIn']) )))
    
    output_graph.add(  (URIRef(uri+str(row.CourseInstID)),
                        URIRef(uri+"personID"),
                        URIRef(uri+Literal(row['personID']) )))
                     
```

```python
# AssignedHours

Assigned_Hours.rename(columns={"Course code": "courseCode", "Study Period": "studyPeriod", "Academic Year": "academicYearCourIn",'Teacher Id': "personID",'Hours':"assignedHours",'Course Instance':"CourseInstID"}, inplace=True)

Assigned_Hours.head()

for index, row in Assigned_Hours.iterrows():

    # make a triple with the object as uri
    output_graph.add(  (URIRef(uri+row.personID),
                        URIRef(RDF.type),
                        URIRef(uri+Literal('Assignment') )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"courseCode"),
                        URIRef(uri+Literal(row['courseCode']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"studyPeriod"),
                        URIRef(uri+Literal(row['studyPeriod']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"academicYearCourIn"),
                        URIRef(uri+Literal(row['academicYearCourIn']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"assignedHours"),
                        URIRef(uri+Literal(row['assignedHours']) )))
    
    output_graph.add(  (URIRef(uri+str(row.personID)),
                        URIRef(uri+"CourseInstID"),
                        URIRef(uri+"CourseInstID") ))
                     
```

```python
output_graph.serialize('mycsv2rdf.ttl',format='turtle')
```
